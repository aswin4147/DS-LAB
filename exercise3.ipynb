```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

1. Draw a line in a diagram from position (1,3) to (2,10) then to (6, 12) and finally to position (18,20). (Mark each point with a beautiful green color and set line color to red and line style dotted)

```python
x = np.array([1,2,6,18])
y = np.array([3,10,12,20])

plt.plot(x,y,marker=".", color="red", ms="10", mec = "green", mfc = "green", ls = "dotted")
plt.show()
```

2. Draw a plot for the following data:

```python
data = {
    "x" : [12, 14, 16, 18, 20, 22, 24],
    "y" : [100, 200, 250, 400, 300, 450, 500]
}

df = pd.DataFrame(data)

sns.scatterplot(x="x", y="y", data=df)

plt.xlabel("Temperature in degree Celsius")
plt.ylabel("Sales")
plt.show()
```

3. Write a python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title

```python
data =np.loadtxt("file.txt", dtype=int)

plt.plot(data[0], data[1])
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("This is the title")
plt.show()
```
