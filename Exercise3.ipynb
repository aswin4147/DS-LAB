```python
bmi = np.array([17,18,19,20])
age = np.array([16,18,24,32])
plt.scatter(bmi, age)
plt.title("CHART")
plt.xlabel("bmi")
plt.ylabel("age")
plt.show()
```

```python
sugar = np.array(['yes', 'no', 'yes', 'no'])
sns.scatterplot(x = bmi, y = age, hue = sugar, palette={'yes':'red', 'no':'green'})
```
