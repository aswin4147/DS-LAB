```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("dtree.csv")
x = data[['sore_throat', 'fever', 'swollen_glands', 'congestion', 'headache']]
y = data[['diagnosis']]
fname = ['sore_throat', 'fever', 'swollen_glands', 'congestion', 'headache']
tname = ['Allergy','Cold','Strep_Throat']
print(data)
```

```python
clf = DecisionTreeClassifier(criterion="entropy")
clf.fit(x,y)
newcase_values = [[0,0,1,0,0]]
newcase_df = pd.DataFrame(newcase_values, columns=fname)
y_new = clf.predict(newcase_df)
plt.figure()
plot_tree(clf, fontsize=10, filled=True, rounded=True, class_names=tname, feature_names=fname)
plt.show()
print("Classification for [0,0,1,0,0] is: ", y_new)
```
